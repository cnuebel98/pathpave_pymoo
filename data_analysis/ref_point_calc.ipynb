```python
import os
import pandas as pd # type: ignore
import ast

# Define the root folder path
root_folder = '/Users/carlonue/Documents/Venvs/pathpave_pymoo/log_all_final/radialGradientMap'

# Initialize an empty list to store all objective values
combined_objective_values = []

# Iterate through all subfolders
for subdir, _, files in os.walk(root_folder):
    for file in files:
        # Check if the file is the CSV we are looking for
        if file == 'optLog.csv':
            file_path = os.path.join(subdir, file)
            
            # Read the first row of the CSV file
            df = pd.read_csv(file_path, nrows=1)
            
            # Access the "objectiveValues" column and convert the string to a list of lists
            objective_values_str = df.loc[0, ' objectiveValues']
            objective_values_list = ast.literal_eval(objective_values_str)
            
            # Append the list of lists to the combined list
            combined_objective_values.extend(objective_values_list)

            # Display the result
            print(f"File: {file_path}")
            print(f"Objective Values (as list of lists): {objective_values_list}")
            print(type(objective_values_list[0][0]))
            print()
```

```python
import matplotlib.pyplot as plt
```

```python
# Separate the combined list into two lists: x and y
x_values = [item[0] for item in combined_objective_values]
y_values = [item[1] for item in combined_objective_values]

# Plotting the values
plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_values, color='blue', marker='o')
plt.title('Objective Values Scatter Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid(True)
plt.show()
```

```python
# Find the worst (minimum and maximum) x and y values
worst_x_max = max(x_values)
worst_y_max = max(y_values)

# Display the results
print(f"Worst (Maximum) X Value: {worst_x_max}")
print(f"Worst (Maximum) Y Value: {worst_y_max}")

print(f"Ref_point = {worst_x_max*1.1, 1.1*worst_y_max}")
```
