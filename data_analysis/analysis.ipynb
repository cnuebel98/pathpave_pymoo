```python
import pandas as pd # type: ignore
import ast
import matplotlib.pyplot as plt # type: ignore
```

```python
df = pd.read_csv('/Users/carlonue/Documents/Venvs/pathpave_pymoo/csv_formatted/gradientObstacles_51_51_AGEMOEA_TwoPointCrossover_RadiusSamplingMutation_100_100000_RandomSampling_PathRepair_0_42_formatted.csv')
```

```python
df[" objectiveValues"] = df[" objectiveValues"].apply(ast.literal_eval) 
```

```python
df[" paths"] = df[" paths"].apply(ast.literal_eval) # 40 sec for one file
```

```python
df[" objectiveValues"][999]
```

```python
# List of indices you want to plot
indices = [49, 99, 199, 299, 399, 499, 599, 699, 799, 899, 999]
colors = ['black', 'red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

# Iterate over the indices and colors
for i, idx in enumerate(indices):
    # Extract the list from the DataFrame at the specified index
    data = df[" objectiveValues"][idx]
    
    # Extracting x and y values
    x_values = [item[0] for item in data]
    y_values = [item[1] for item in data]
    
    # Plotting the data
    plt.scatter(x_values, y_values, color=colors[i])

# Labeling the axes
plt.xlabel('Steps')
plt.ylabel('Weight')

# Adding a title (optional)
plt.title('Steps vs. Weight')

# Adding a legend to differentiate the sets
plt.legend()

# Display the plot
plt.show()
```
